# **Group By**

The groupby() function in pandas is used to group data based on one or more columns  in DataFrame

```python
df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
```

```python
S_Qunatity= df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
sns.barplot(x='Size' , y = 'Quantity' , data=S_Qunatity)
```

**Note:From above graph you can see the most of the qty buys M-Size in the sales**
